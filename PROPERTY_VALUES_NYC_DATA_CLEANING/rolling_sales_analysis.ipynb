```python
# Only take columns relevant to sale price

def drop_cols(borough:pd.DataFrame) -> pd.DataFrame:
       return borough.drop(['BLOCK', 'LOT', 'EASEMENT',
       'BUILDING CLASS AT PRESENT', 'APARTMENT NUMBER', 'TAX CLASS AT PRESENT',
       'RESIDENTIAL UNITS', 'COMMERCIAL UNITS', 'TOTAL UNITS', 'BUILDING CLASS CATEGORY',
       'TAX CLASS AT TIME OF SALE', 'BUILDING CLASS AT TIME OF SALE'], axis='columns')

cleaned_bronx = drop_cols(bronx)
cleaned_brooklyn = drop_cols(brooklyn)
cleaned_queens = drop_cols(queens)
cleaned_statinisland = drop_cols(statinisland)
cleaned_manhattan = drop_cols(manhattan)
```

```python
# Merge all datasets into one large city dataset
nyc = pd.concat([cleaned_manhattan, cleaned_bronx], ignore_index=True)
nyc = pd.concat([nyc, cleaned_brooklyn], ignore_index=True)
nyc = pd.concat([nyc, cleaned_queens], ignore_index=True)
nyc = pd.concat([nyc, cleaned_statinisland], ignore_index=True)
nyc
```

```python
# Let's fix column names
nyc.columns = nyc.columns.str.strip()
nyc.columns = nyc.columns.str.replace(' ', '_')
nyc
```

```python
# Remove commas from land_square_feet, gross and sale_price
def replace_vals(arr:list) -> list:
    return [str(i).replace(',', '') for i in arr]
nyc.LAND_SQUARE_FEET = replace_vals(nyc.LAND_SQUARE_FEET)
nyc.GROSS_SQUARE_FEET = replace_vals(nyc.GROSS_SQUARE_FEET)
nyc.SALE_PRICE = replace_vals(nyc.SALE_PRICE)
nyc
```

```python
# Converting necessary rows to numeric values
cols = ['BOROUGH', 'ZIP_CODE', 'LAND_SQUARE_FEET', 'GROSS_SQUARE_FEET', 'YEAR_BUILT', 'SALE_PRICE']
nyc[cols] = nyc[cols].apply(pd.to_numeric, errors='coerce')
nyc
```

```python
# Take all rows where sale price is not 0
nyc = nyc[nyc.SALE_PRICE != 0]
nyc
```

```python
nyc.YEAR_BUILT.isna().sum()
```

```python
# Year built is an expected large factor in determining price, so we can remove any rows that are NaN
nyc = nyc[nyc.YEAR_BUILT.notna()]
nyc
```

```python
# check the min sale_price
nyc.SALE_PRICE.min()
```

```python
# check the max sale_price
nyc.SALE_PRICE.max()
```

```python
# we can assume a sale price of under 100,000 is actually a rental of some sort so we can remove these values
# Check how many houses have reported under $100,000
nyc = nyc[nyc.SALE_PRICE > 100_000]
nyc
```

```python
nyc.ZIP_CODE, nyc.YEAR_BUILT = nyc.ZIP_CODE.astype(int), nyc.YEAR_BUILT.astype(int)
nyc
```

```python
# We can also assume a property with over 2500 gross square feet and under 200,000 price is a rental property, even if it is not listed as an apartment
# NOTICE: NOT ALL SALES LISTED AS RENTAL UNDER BUILDING CLASS ARE ACTUALLY BEING RENTED SO WE HAVE TO BE CAREFUL AS TO WHAT WE ARE REMOVING

nyc.drop(nyc[(nyc.GROSS_SQUARE_FEET > 2500) & (nyc.SALE_PRICE < 200_000)].index, inplace=True)
nyc
```

```python
nyc.SALE_DATE.value_counts() # This data is farely recent, from 2021 to 2022
```

```python
# We need to change the sale date from string to date value, we have to format it as yyyy
nyc.SALE_DATE = pd.to_datetime(nyc.SALE_DATE)

def get_year(arr:list) -> list:
    return [str(i)[:4] for i in arr]
    
nyc.SALE_DATE = get_year(nyc.SALE_DATE)
nyc
```

```python
# Let's rearrange the column order for better readability
nyc = nyc[['BOROUGH', 'ZIP_CODE', 'YEAR_BUILT',
'LAND_SQUARE_FEET', 'GROSS_SQUARE_FEET',
'SALE_PRICE', 'SALE_DATE', 'NEIGHBORHOOD', 'ADDRESS']]

nyc
```

```python
# Check for NaNs one more time
nyc.isna().sum()
```

```python
# Because Gross_Square_feet is our most important feature for predicting values, we must remove any rows missing this value
nyc = nyc[nyc.GROSS_SQUARE_FEET.notna()]
nyc
```

```python
# Let's take numerical rows to create our learning model. We have to keep address and neighborhood for our dashboard purposes 
# We only need Gross_Square_Feet and not Land_Square_Feet

nyc_numerical = nyc.drop(['ADDRESS', 'LAND_SQUARE_FEET', 'ZIP_CODE'], axis='columns')
nyc_numerical
```

```python
# Check the correlations of our model
nyc_numerical.corr() # We can see that gross square feet has the highest correlation to sale price, zipcode and borough having a negative correlation 
```

## BUILDING THE MODEL

We will use a decision tree because of the many factors that determine sale price.

```python
inputs = nyc_numerical.drop('SALE_PRICE', axis='columns') # Our independent values
inputs
```

```python
targets = nyc_numerical.SALE_PRICE # our dependant variable
targets
```

```python
# Let's Preprocess our data
from sklearn.preprocessing import LabelEncoder
neighborhood_le = LabelEncoder()
inputs['NEIGHBORHOOD_N'] = neighborhood_le.fit_transform(inputs.NEIGHBORHOOD)
inputs
```

```python
# let's create a separate dataframe to keep track of our neighborhoods from our processed inputs and dataframe
neighborhood_tracker = pd.DataFrame()
neighborhood_tracker['NEIGHBORHOOD'], neighborhood_tracker['ENCODED_VALUE'] = inputs['NEIGHBORHOOD'], inputs['NEIGHBORHOOD_N']
neighborhood_tracker.to_csv('encoded_tracker.csv') # Let's keep track of it in a file
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(inputs, targets)
```

```python
# Let's check accuracy of model
model.score(inputs, targets) # 90.6%, a very accurate score
```

```python
# Let's check model by using values we know from our first row in our dataset for accuracy 
print(model.predict([[1, 2003, 2160, 2021, 1]])[0])
```

```python
# Let's predict prices for a house next year in Bronx and the total sq ft of the house is 2200, it was built in 1980, and the neighborhood is whitestone
# Format is Borough #, YEAR_BUILT, SQ_FEET TOTAL, SALE_DATE, NEIGHBORHOOD
print(model.predict([[2, 1980, 2200, 2023, 235]])[0]) # Predicts at $810,000 which is pretty reasonable for house being sold in 2023 built in 1980 at that size
```
